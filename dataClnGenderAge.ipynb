```python
df_ageGroup.head()
```

```python
df_ageGroup.dropna(axis = 1, inplace=True)
df_ageGroup.head()
```

```python
df_ageGroup.drop(df_ageGroup[['Code', 'Klassifikator', 'Klassifikator_ru','Klassifikator_en']], axis = 1, inplace=True)
```

```python
df_ageGroup.rename(index={0 : 'Babises(0-2)', 1 : '6-7', 2:'8-15', 3: '16-17', 4: '18-19', 5: '20-24', 6: '25-29', 7: '30-34', 8: '35-39', 9:'40-49', 10: '50-59', 11: '60-64', 12 : '65+' }, inplace=True)
```

```python
df_ageGroup.head(15)
```

```python
sum_of_children = df_ageGroup.loc[['6-7', '8-15']].sum()
sum_of_youngAdults = df_ageGroup.loc[['16-17', '18-19', '20-24', '25-29']].sum()
sum_of_middleAgedAdults = df_ageGroup.loc[['30-34', '35-39', '40-49']].sum()
sum_of_old = df_ageGroup.loc[['50-59', '60-64', '65+']].sum()

```

```python
df_ageGroup.loc['Children(6-15)'] = sum_of_children
df_ageGroup.loc['Young Adults(16-29)'] = sum_of_youngAdults
df_ageGroup.loc['Middle-aged Adults(30-49)'] = sum_of_middleAgedAdults
df_ageGroup.loc['Old Adults(50+)'] = sum_of_old
```
